```python
import pandas as pd
import numpy as np
from scipy import sparse
from sklearn.metrics.pairwise import cosine_similarity

movie=pd.read_csv('movies1.csv')
rating=pd.read_csv('ratings.csv')
rating=pd.merge(movie,rating)
user_rating=rating.pivot_table(index=['userId'],columns=['title'],values=['rating'])
user_rating=user_rating.dropna(thresh=10,axis=1).fillna(0)

def standardize(row):
    new_row=(row-row.mean())/(row.max()-row.min())
    return new_row

df1=user_rating.apply(standardize)
```

```python
new_df=cosine_similarity(df1.T)
new_df
```

```python
main_df=pd.DataFrame(new_df,index=df1.columns,columns=df1.columns)
main_df.head()
```

```python
def myfun(movie_name,user_rating):
    print("vladivostok:",movie_name)
    similar_score=main_df[movie_name]*(user_rating-2.5)
    similar_score=similar_score.sort_values(ascending=False)
    return similar_score
myfun("burbs, The (1989)",3.5)
```
